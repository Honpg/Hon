### Calculate slope based on lat,lon,elevation


```python
import pandas as pd
from math import sin, cos, sqrt, atan2, radians, atan, pow
import random

# Haversine formula for distance between 2 coordinates latitude and longitude
lst1 = []
lst2 = []

slope_degrees = [0]
R = 6373000

df = pd.read_csv("ndvi_elevation.csv")
for row in df.values:
    lst1.append([row[0], row[1], row[2]])
while max(slope_degrees) <= 40:
    for i in lst1:
        
        remaining_data = [pair for pair in lst1 if pair != i]
        random_pairs = random.sample(remaining_data, 4)
        
        lat = i[0]
        long = i[1]
        ele_of_i = i[2]
        
        a = random_pairs[0]
        b = random_pairs[1]
        c = random_pairs[2]
        d = random_pairs[3]
        
        ele_of_a = a[2]
        ele_of_b = b[2]
        ele_of_c = c[2]
        ele_of_d = d[2]
        
        slat1, slon1, elat1, elon1 = radians(float(a[0])), radians(float(a[1])), radians(float(lat)), radians(float(long))

        dlon1 = elon1 - slon1
        dlat1 = elat1 - slat1
        
        a_1 = sin(dlat1 / 2)**2 + cos(slat1) * cos(elat1) * sin(dlon1 / 2)**2
        c1 = 2 * atan2(sqrt(a_1), sqrt(1 - a_1))

        distance_rigth_long = R * c1
        ######################################################################################################################
        slat2, slon2, elat2, elon2 = radians(float(b[0])), radians(float(b[1])), radians(float(lat)), radians(float(long))

        dlon2 = elon2 - slon2
        dlat2 = elat2 - slat2
        
        a_2 = sin(dlat2 / 2)**2 + cos(slat2) * cos(elat2) * sin(dlon2 / 2)**2
        c2 = 2 * atan2(sqrt(a_2), sqrt(1 - a_2))

        distance_left_long = R * c2
        ##########################################################################################################################
        slat3, slon3, elat3, elon3 = radians(float(c[0])), radians(float(c[1])), radians(float(lat)), radians(float(long))

        dlon3 = elon3 - slon3
        dlat3 = elat3 - slat3
        
        a_3 = sin(dlat3 / 2)**2 + cos(slat3) * cos(elat3) * sin(dlon3 / 2)**2
        c3 = 2 * atan2(sqrt(a_3), sqrt(1 - a_3))

        distance_up_lat = R * c3
        ##########################################################################################################################
        slat4, slon4, elat4, elon4 = radians(float(d[0])), radians(float(d[1])), radians(float(lat)), radians(float(long))

        dlon4 = elon4 - slon4
        dlat4 = elat4 - slat4
        
        a_4 = sin(dlat4 / 2)**2 + cos(slat4) * cos(elat4) * sin(dlon4 / 2)**2
        c4 = 2 * atan2(sqrt(a_4), sqrt(1 - a_4))

        distance_down_lat = R * c4
        
        
        dz_dx = (ele_of_a - ele_of_b) / (distance_rigth_long + distance_left_long)
        dz_dy = (ele_of_c - ele_of_d) / (distance_up_lat + distance_down_lat)
        
        rise_run = sqrt(pow(dz_dx, 2) + pow(dz_dy, 2))
        
        slope = atan(rise_run) * 57.29578
        
        slope_degrees.append(slope)
    
    
    
df = pd.read_csv('ndvi_elevation.csv')

df['slope'] = slope_degrees

df.to_csv('last_check.csv', index = False)
```
